```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.arange(1,11)
```

```python
a
```

```python
b=a**2
```

```python
b
```

```python
plt.plot(a,b,'r--')
```

```python
plt.xlabel('Numbers')
plt.ylabel('Squares of Numbers')
plt.title('My First Plot')
plt.plot(a,b)
```

```python
plt.plot(b,a,'r*-')   #(b,a,g*-)
plt.xlabel('Squares')
plt.ylabel('Numbers')
plt.title('Second Plot')
```

```python
plt.subplot(1,2,1)  #(rows,col,plot number)  # combining both plot in a grid  # Using Subplot
plt.plot(a,b,'r--')
plt.subplot(1,2,2)
plt.plot(b,a)
```

```python
fig=plt.figure()
```

```python
####FRAME WITHIN FRAME  #### Using Figure Object
fig=plt.figure()   # creating frame
ax1=fig.add_axes([0.1,0,0.9,0.9]) # adding border([left margin,distance from bottom,total width of frame,height of frame])
ax2=fig.add_axes([0.2,0.5,0.4,0.3])
ax1.plot(a,b)# outer frame is not visible
ax1.set_xlabel('numbers')
ax1.set_ylabel('squares')
ax2.plot(b,a)
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,0.4,1]) # adding border([left margin,distance from bottom,total width of frame,height of frame])
ax2=fig.add_axes([0.5,0,0.4,1])
ax1.plot(a,b)# outer frame is not visible
ax1.set_xlabel('numbers')
ax1.set_ylabel('squares')
ax2.plot(b,a)
ax2.set_ylabel('numbers')
ax2.set_xlabel('squares')
```

```python
fig,axes=plt.subplots(1,2) # Same using figure object and subplots
axes[0].plot(a,b)
axes[1].plot(b,a)
```

```python
fig,axes=plt.subplots(1,4)
axes[0].plot(a,b)
axes[1].plot(b,a)
axes[2].plot(a,a)
```

```python
plt.plot(a,b)   # 1.plot without object
```

```python
fig=plt.figure() #### 2. With Figure obj(Customizable) 
ax1=fig.add_axes([0,0,1,1])
ax1.plot(a,b)
```

```python
fig,axes=plt.subplots() #####   3. with figure obj and subplots (Not customisable) (bydefualt axes(1,1))
axes.plot(a,b)
```

```python
fig,axes=plt.subplots(figsize=(10,3))    #######fig size to set size of frame
```

```python
fig,axes=plt.subplots(figsize=(10,5))
axes.plot(a,b)
axes.set_xlabel('numbers')
axes.set_ylabel('squares')
axes.set_title('size')
fig.savefig('E:\save.png')   ##########saving figure
```

```python
fig,axes=plt.subplots(1,3)   ### plots in same frame , x axis must be same in both plots
axes[0].plot(a,b,'b*-',label="x**2")
axes[0].plot(a,c,'b*-',label="x**3")
plt.legend(loc=2)

axes[1].plot(a,b,a,c)
axes[1].set_xlim([4,6])  ######zooming in in 4 to 6 range on x axis
axes[1].set_ylim([0,200]) ########zooming in in 0 to 200 on y axis

axes[2].plot(a,b,'b*-',label="x**2")
axes[2].plot(a,c,'r--',label="x**3")
plt.legend(loc=2)
axes[2].set_xlim([4,6])
axes[2].set_ylim([0,200])
    
```
